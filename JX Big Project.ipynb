```python
#导入需要使用的库
import numpy as np
import pandas as pd #读取csv文件的库
import os,sys
import fnmatch
import matplotlib
from bokeh.plotting import figure, output_file, show,Column,Row
from bokeh.models import ColumnDataSource
from bokeh.palettes import viridis
Color=viridis
from bokeh.io import export_svgs
```

```python
def file_read(file_path,min_value,max_value,col_name):
    name_list=os.listdir(file_path)
    domain = os.path.abspath(file_path) #获取文件夹的路径
    memory_mean={}

    for name in name_list:#文件读取

        info = os.path.join(domain,name) #将路径与文件名结合起来就是每个文件的完整路径
        data = pd.read_csv(info)
           
        #数据处理
            
        dat=data[(data['x']>=min_value)&(data['x']<=max_value)] #按照条件删选data中的值，将符合条件的放置到dat中
        m=dat["y"].mean() #对dat中y列取平均值，并将平均值保存到变量m中
        memory_mean[name]=m #将m的数值放入到字典中，索引设置为文件名name

    #data['name'].str.split('|',expand=True)
    #df=pd.DataFrame([memory_mean])
    df = pd.DataFrame(pd.Series(memory_mean),columns=['vd_ave']) #将字典格式的memory_mean转换为dataframe格式，并命名数值一列为‘value’
    df = df.reset_index().rename(columns={'index':'name'}) #将‘index’命名Wie‘name’
    sdf=df['name'].str.split('_',expand=True) #将'name'列以‘_’进行拆分，并存入sdf中
    sdf.columns=col_name #

    sheet=pd.concat([df,sdf],axis=1)
    sheet=sheet.apply(pd.to_numeric, errors='ignore')
    sheet=sheet.sort_values(by='Iamp')
    
    return(sheet)
  
```

```python
in_path='SOI_2D_B_x'  #自定义输入路径
out_path='XXX'    #自定义图片和csv的输出路径
out_name='XXX'    #给自己的这个项目取个名字
min_value=15e-9
max_value=80e-9
col_name=['Iamp','x','y','L','Temp','Vg','csv']   #这个对应文件的命名格式，多了自己往后添加都行
sheet=file_read(in_path,min_value,max_value,col_name)
pd.DataFrame(sheet).to_csv( out_path+out_name+"-"+'sheet.csv')
```

```python
markers=["o","v","*","+","x","^",'hex','square','diamond',"o","v","*","+","x","^",'hex','square','diamond',] 
for name in col_name:
    if (name != 'Iamp')&(name != 'csv'):
         
        vals=sheet[name].drop_duplicates().values.tolist()
        names=locals()
        p=figure(plot_width=800, plot_height=600) #这里的图高度宽度可以自己改
        mp=figure()
        i=0
        
        size=len(vals)+1
        if size >2:
            for val in vals:  
                i=i+1       
                names['p' + str(val) ] = figure(plot_width=800, plot_height=600) #这个是那几个分图，高度宽度可以自己改
                x=sheet[(sheet[name]==val)]['vd_ave']
                y=sheet[(sheet[name]==val)]['Iamp']
                dat=sheet[(sheet[name]==val)]
                pd.DataFrame(dat).to_csv( out_path+out_name+"-"+str(name)+"-"+str(val)+'sheet.csv')
                p.line(x, y,  line_width=2, legend=str(name)+"-"+str(val),color=Color(size)[i])# #线条长度与大小可以自己改，下同  
                p.scatter(x, y,  size=10,legend=str(name)+"-"+str(val),marker=markers[i-1],color=Color(size)[i])# size点的改大小
                names['p' + str(val) ] .scatter(x, y,  size=10,legend=str(name)+"-"+str(val),marker=markers[9],color=Color(size)[i])#color=Color[vg]
                names['p' + str(val) ] .line(x, y, line_width=2,legend=str(name)+"-"+str(val),color=Color(size)[i])#
                mp=Row(names['p' + str(val) ],mp)
                names['p' + str(val)].output_backend="svg"
                export_svgs(names['p' + str(val)],out_path+out_name+"-"+str(name)+"-"+str(val)+".svg")    
            mp=Column(p,mp)
            show(mp)
            output_file(out_path+out_name+"-"+str(name)+".html")
            p.output_backend="svg"   
            export_svgs(p,out_path+out_name+"-"+str(name)+".svg") 
            
```
